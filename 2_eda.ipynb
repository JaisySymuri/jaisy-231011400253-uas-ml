### EDA (Exploratory Data Analysis)

EDA (**Exploratory Data Analysis**) adalah tahap awal dalam analisis data untuk **memahami isi dan karakteristik dataset** sebelum dibuat model.

Tujuan utama EDA adalah untuk:

1. **Melihat struktur data** — jenis kolom, tipe data, jumlah baris, dan apakah ada data kosong.  
2. **Memahami pola dan hubungan antar variabel** — misalnya, apakah jam belajar berpengaruh terhadap nilai akhir.  
3. **Menemukan anomali atau outlier** — seperti nilai yang tidak wajar.  
4. **Membuat visualisasi** — grafik, histogram, dan heatmap untuk memahami data secara intuitif.

Dengan EDA, kita dapat mengetahui kondisi data sebelum melakukan preprocessing atau modeling, sehingga model yang dibuat lebih **akurat, relevan, dan terpercaya**.

```python
import pandas as pd

df = pd.read_csv('1-student_performance.csv')
```

2. Info dasar dataset

```python
print("Shape:", df.shape)

df.info()
df.describe()
df.isnull().sum()
```

3. Overview kolom kategori

```python
cat_cols = df.select_dtypes(include=['object']).columns
for col in cat_cols:
    print(f"\n{col} → {df[col].unique()}")
```

4. Explorasi Visual

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid", palette="pastel")
```

4.1 - Distribusi Target

```python
plt.figure(figsize=(6,4))
sns.histplot(x=df['G3'], kde=True, bins=10)   
plt.title('Distribusi Nilai G3')
plt.show()
```

4.2 - Gender vs Nilai G3

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='sex', y='G3', data=df)
plt.title('Gender vs Nilai G3')
plt.show()
```

4.3 - Heatmap Korelasi

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(numeric_only=True), annot=True, fmt=".2f", cmap="coolwarm")
plt.title('Heatmap Korelasi')
plt.show()
```

4.4 - Waktu Belajar vs Nilai G3

```python
sns.barplot(x='studytime', y='G3', data=df)
plt.title('Waktu Belajar vs Nilai G3')
plt.show()
```

4.5 - Absen vs Nilai G3

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x='absences', y='G3', data=df)
plt.title('Absen vs Nilai')
plt.show()
```

5. Hubungan antar nilai kelas 10 dan 11 (G1, G2) dan nilai kelas 12 (G3)

```python
sns.pairplot(df[['G1','G2','G3']])
plt.suptitle("Hubungan antar nilai kelas 10 dan 11 (G1, G2) dan nilai kelas 12 (G3)", y=1.02)
plt.show()

df[['G1','G2','G3']].corr()
```

6. Kesimpulan Awal

```python
print("Rata-rata nilai G3 berdasarkan akses internet:")
print(df.groupby('internet')['G3'].mean())

print("\nRata-rata nilai G3 berdasarkan dukungan keluarga:")
print(df.groupby('famsup')['G3'].mean())
```
